```python
from ema_workbench import load_results
import pandas as pd


experiments_p, outcomes_p = load_results('../results/rfr_results.tar.gz')
experiments_bc, outcomes_bc = load_results('../results/basecase_results.tar.gz')
experiments_d5, outcomes_d5 = load_results('../results/dike_5_results.tar.gz')
outcomes_p_df = pd.DataFrame.from_dict(outcomes_p)
outcomes_p_df['policy'] = 'rfr'
outcomes_bc_df = pd.DataFrame.from_dict(outcomes_bc)
outcomes_bc_df['policy'] = 'bc'
outcomes_d5_df = pd.DataFrame.from_dict(outcomes_d5)
outcomes_d5_df['policy'] = 'd5'

outcomes_df = pd.concat([outcomes_p_df, outcomes_bc_df, outcomes_d5_df])
```

```python
outcomes_df
```

```python
locations = ["A.1", "A.2", "A.3", "A.4", "A.5"]
stages = range(0,3)
df_locations_agg = pd.DataFrame()
kpi = ['Expected Annual Damage','Dike Investment Costs', 'Expected Number of Deaths']

for kpi in kpi:
    for stage in stages:
        columns = []
        for location in locations:
            columns.append(f'{location}_{kpi} {stage}')

        df_locations_agg[f'Total {kpi} {stage}'] = outcomes_df[columns].sum(axis=1)

        
kpi_2 = ['RfR Total Costs', 'Expected Evacuation Costs']

for kpi in kpi_2:
    for stage in stages:
        df_locations_agg[f'{kpi} {stage}'] = outcomes_df[f'{kpi} {stage}']
        
```

```python
df_locations_agg['policy'] = outcomes_df['policy']
```
