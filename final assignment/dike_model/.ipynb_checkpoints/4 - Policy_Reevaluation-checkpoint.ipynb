# 4. Re-evaluation under deep uncertainty

Optimizing generates a set of promising strategies that illustrate the key trade-offs on the relevant objectives (J.H.Kwakkel, Walker, et al. 2016). First, to counter scenario dependency that might result from optimizing policies only on the five previously selected worst-case scenarios, the identified policies were tested over the various deeply uncertain factors to assess their robustness against uncertainties. A larger scenario sampling also allows us to obtain a more representative value of the robustness of each policy. Due to time constraints it was kept to 500 scenarios. Second, the domain criterion robustness function was defined and the following outcomes were evaluated under the metric:

- Expected Number of Deaths
- Dike Investment Costs
- RfR investment Costs  

Expected Annual Damage and Evacuation costs were not evaluated under the domain criterion because no other stakeholder had set a minimum threshold. Given that the client's priority is to maximise RfR investment and minimize Dike Investment, there was no motivation to assess the robustness of those outcomes. Third, 4 policies which met the criteria to potentially be approved by Rijkswatersaat or the Delta Commision were selected and their trade-offs analysed.

To meet the criterion specified by the Delta COmmision, we  only chose the policies that keep the number of deaths under the threshold in ALL scenarios

```python
outcomes_reev_df['policy']= experiments_reev['policy']
policy_names = outcomes_reev_df.policy.unique()

max_num_deaths = 0.0001 * 3 * 5

# Returns a list with the policy name if it meets the maximum number of deaths criterion under all scenarios
policies_of_interest = []
for policy in policy_names:
    policy_df = outcomes_reev_df[outcomes_reev_df['policy'] == policy]
    policy_df_logical = policy_df['Expected Number of Deaths'] > max_num_deaths
    type(policy_df_logical)
    if policy_df_logical.any():
        continue
    else:
        policies_of_interest.append(policy)

policies_of_interest
```

We can see that there is no policy that obeys the threshold under all scenarios. We then proceed to calculate the domain robustness using the thresholds specified by Rijkswaterstaat and the Delta Commision during the debate. These are:
- Rijkswatersaat:
    * Maximum RfR Investment Costs < 1.1e9 euros
    * Maximum Dike Investment Costs < 3.04e8 euros
- Delta Commission:
    * Maximum Expected Number of Deaths < 0.0001 * 3 * 5
    
The objective is to find a policy that satifies this requirements at the same time that looks for the interest of the client, i.e. maximise RfR investment and minimize Dike Investment.

Next, we define the domain criterion robustness function that evaluates Expected Number of Deaths, Dike Investment Costs and, RfR investment Costs. Additionally, we create a dummy function to evaluate the Expected Annual Damage and the Evacuation Costs. The purpose of using a dummy function is to facilitate the posterior visualisation making.

```python
import functools

# Domain criterion robustness metric function
def robustness(direction, threshold, data):
    if direction == 'SMALLER':
        return np.sum(data<=threshold)/data.shape[0]
    else:
        return np.sum(data>=threshold)/data.shape[0]

# Dummy function that facilitates the posterior analysis
def dummy(data):
    return data[0]
    
SMALLER = 'SMALLER'


max_num_deaths = 0.0001 * 3 * 5
max_rfr_costs = 1.1e9
max_dike_costs = 3.04e8

Expected_Annual_Damage = dummy
Dike_Investment_Costs = functools.partial(robustness, SMALLER, max_dike_costs)
Expected_Number_of_Deaths = functools.partial(robustness, SMALLER, max_num_deaths)
RfR_Investment_Costs = functools.partial(robustness, SMALLER, max_rfr_costs)
Evacuation_Costs = dummy
```

Now that the functions are defined, we proceed to evaluate the variables. To get a first impression of the outcome, we make a parcoord with the robustness values of the 42 policies.

```python
from ema_workbench.analysis import parcoords
import matplotlib.lines as mlines


funcs = {'Expected Annual Damage': Expected_Annual_Damage,
         'Dike Investment Costs': Dike_Investment_Costs,
         'Expected Number of Deaths':Expected_Number_of_Deaths,
         'RfR Investment Costs':RfR_Investment_Costs,       
         'Evacuation Costs': Evacuation_Costs}

total_scores = {}
for policy in experiments_reev.policy.unique():
    scores = {}
    logical = experiments_reev['policy'] == policy
    
    temp_outcomes = {k:v[logical] for k,v in outcomes_reev.items()}
    
    for k, v in temp_outcomes.items():
        score = funcs[k](v)
        scores[k] = score
    total_scores[policy] = scores

data_domain = pd.DataFrame(total_scores).T
data_domain = data_domain.drop(data_domain.columns[0], axis=1)
data_domain = data_domain.drop(data_domain.columns[-2], axis=1)
limits = parcoords.get_limits(data_domain)


paraxes = parcoords.ParallelAxes(limits)
paraxes.plot(data_domain)
paraxes.invert_axis('Dike Investment Costs')
plt.show()
```

At first sight we can already see that all of the policies meet the criterion specified by Rijkswatersaat on the Dike Investment Costs. Next, we can see that not all policies meet the threshold set on the RfR investment costs. This result is probably due to minimizing over the Dike Investment Costs but not over RfR invesmtent costs. Finally, we can see that there is no policy that meets the threshold on the Expected Number of Deaths under all scenarios.

 To get a better grasp of the results we check the top performers on the robustness of Expected Number of Deaths. As we can see, the top performer does not meet the requirement on the RfR costs. Assuming that the Delta Commission might we willing to bring the threshold down to 0.766 of occasions, we might be facing a potential trade-off.

We plot the outcomes for each policy on each scenario to get some insights on their behavior and the potential trade-offs.

```python
policy1_df = outcomes_reev_df[outcomes_reev_df['policy']==potential_policies[0]].iloc[:,:-1]
policy2_df = outcomes_reev_df[outcomes_reev_df['policy']==potential_policies[1]].iloc[:,:-1]
policy3_df = outcomes_reev_df[outcomes_reev_df['policy']==potential_policies[2]].iloc[:,:-1]
policy4_df = outcomes_reev_df[outcomes_reev_df['policy']==potential_policies[3]].iloc[:,:-1]

frames = [policy1_df,policy2_df,policy3_df,policy4_df]
all_data = pd.concat(frames)

limits = parcoords.get_limits(all_data)
axes = parcoords.ParallelAxes(limits)
axes.plot(policy1_df, color=sns.color_palette()[0], label='Policy 1')
axes.plot(policy4_df, color=sns.color_palette()[3], label='Policy 2')
axes.plot(policy2_df, color=sns.color_palette()[1], label='Policy 3')
axes.plot(policy3_df, color=sns.color_palette()[2], label='Policy 4')

plt.rcParams.update(plt.rcParamsDefault)
plt.rcParams["figure.figsize"] = (10,4)

axes.legend()
plt.savefig('../images/parcoord_500.png', dpi=300, bbox_inches = "tight")
plt.show()
```

Putting all KPIs side-by-side, it can be observed that the policies do present certain trends in trade-offs between the performance on KPIs even though there are influences of the different scenarios. Policy 3 and 4 seem to make the same trade-offs in which lower policy costs result in slightly more deaths and particularly more damage costs.
