### Base case comparison and policy exploration ###

In the previous notebook "1 - Exploration" we explored the scenario space under the 'base case' policy. In this notebook we will explore the policy space and compare the results with the 'base case' policy to see whether any policy will improve the current situation.

```python
# Import libraries
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import networkx as nx

# Import the workbench and model
from ema_workbench import (Model, CategoricalParameter, ScalarOutcome, IntegerParameter, RealParameter)
from ema_workbench import (MultiprocessingEvaluator, Policy, Scenario)
from ema_workbench.em_framework.evaluators import perform_experiments
from ema_workbench.em_framework.samplers import sample_uncertainties
from ema_workbench.util import ema_logging, utilities
from dike_model.dike_model_function import DikeNetwork 
from dike_model.problem_formulation import get_model_for_problem_formulation
ema_logging.log_to_stderr(ema_logging.INFO)
```

We run the workbench with 375 scenarios per random generated policy, and we instantiate 100 policies; resulting in 37500 experiments. After running, we load in the results from the policy and base case run since we don't need to run it each time we run the notebook. Afterwards, we merge them in dataframes and append a column policy (which is random or basecase), so we can distinct the outcomes and more easily visualize them.

```python
#This part is commented out because the results are loaded from a csv file.
# dike_model, planning_steps = get_model_for_problem_formulation(5)

# def sum_over(*args):
#     return sum(args)

# with MultiprocessingEvaluator(dike_model) as evaluator:
#     results = evaluator.perform_experiments(scenarios=375, policies=100)

# utilities.save_results(results, 'results/375times100PoliciesPF5.csv')

# Load in the results, for both the with and the without policy runs.
experiments, outcomes = utilities.load_results('results/375times100PoliciesPF5.csv')
experiments_basecase, outcomes_basecase = utilities.load_results('results/5000ScenariosNoPolicyPF5.csv')

# Transform the outcomes to a dataframe.
df_outcomes = pd.DataFrame(outcomes)
df_outcomes_basecase = pd.DataFrame(outcomes_basecase)

# Merge the seperate dataframes into one.
experiments_combined = pd.concat([experiments, experiments_basecase], sort = True)
outcomes_combined = pd.concat([df_outcomes, df_outcomes_basecase], sort = True)

# Rename all the runs that include policy to "random" (indicating random policies), and all the others keep their original name of 'base case'
experiments_combined.loc[experiments_combined["policy"] != "base case", "policy"] = "random"

# And add them to a column within the combined outcomes dataframe
policies = experiments_combined["policy"]
outcomes_combined["policy"] = policies
```

With the retrieved "outcomes_combined" dataframe, we can aggregate the variables in order to visualize them against each other.

We implement the aggregate functions 'aggregate_kpi' and 'variance_kpi' created in notebook "1 - Exploration" and calculate the aggregated variables. Note that variance_kpi now *does* include the dike investment costs, as these are now a non-zero number, because random policy has now been implemented.

```python
#This function has been explained in the '1 - Exploration.ipynb' notebook
def aggregate_kpi(data, kpi, aggregate):
    locations = ["A.1", "A.2", "A.3", "A.4", "A.5"]
    kpi_columns = []
    
    if kpi == "RfR Total Costs" or kpi == "Expected Evacuation Costs":
        kpi_columns.append(kpi + " 0")
        kpi_columns.append(kpi + " 1")
        kpi_columns.append(kpi + " 2")
        
        data[kpi] = data[kpi_columns].sum(axis=1)
    else:
        if aggregate is "total":
            for location in locations:
                kpi_columns.append(location + "_" + kpi + " 0")
                kpi_columns.append(location + "_" + kpi + " 1")
                kpi_columns.append(location + "_" + kpi + " 2")

            data[kpi] = data[kpi_columns].sum(axis=1)

        else:
            for location in locations:
                kpi_columns = []
                kpi_columns.append(location + "_" + kpi + " 0")
                kpi_columns.append(location + "_" + kpi + " 1")
                kpi_columns.append(location + "_" + kpi + " 2")

                data[location + "_" + kpi] = data[kpi_columns].sum(axis=1)
                
    return data

#This function is for calculating the variance / standard deviation of location-specific costs over the different locations. 
def variance_kpi(data):
    locations = ["A.1", "A.2", "A.3", "A.4", "A.5"]
    kpi_columns = []
    
    for location in locations:
        kpi_columns.append(location + "_Expected Annual Damage")
        kpi_columns.append(location + "_Dike Investment Costs")
        
    df_merged = pd.DataFrame([data[[kpi_columns[x], kpi_columns[x+1]]].sum(axis=1) for x in range(0, len(kpi_columns), 2)]).T
    
    data["Variance"] = np.std(df_merged, axis=1)
    
    return data


# Append the KPIs we would like to analyse.

outcomes_combined = outcomes_combined.reset_index(drop=True)
outcomes_combined = aggregate_kpi(outcomes_combined, "Expected Number of Deaths", "total")
outcomes_combined = aggregate_kpi(outcomes_combined, "Expected Annual Damage", "total")
outcomes_combined = aggregate_kpi(outcomes_combined, "RfR Total Costs", "total")
outcomes_combined = aggregate_kpi(outcomes_combined, "Expected Evacuation Costs", "total")
outcomes_combined = aggregate_kpi(outcomes_combined, "Dike Investment Costs", "total")
outcomes_combined = aggregate_kpi(outcomes_combined, "Expected Annual Damage", "location")
outcomes_combined = aggregate_kpi(outcomes_combined, "Dike Investment Costs", "location")
outcomes_combined = variance_kpi(outcomes_combined)
```

We create a new dataframe with all the outcomes of interest and cohering policies, since this will be easier to distinct and gives more insight in the values. We also create the KPI "total policy costs", which is the total costs of the policy (RfR costs, Evacuation costs and Dike investment costs per location).

```python
df_ooi = outcomes_combined[["policy", 'Expected Number of Deaths', 'Expected Annual Damage',
       'RfR Total Costs', 'Expected Evacuation Costs', 'Dike Investment Costs',
       'A.1_Expected Annual Damage', 'A.2_Expected Annual Damage',
       'A.3_Expected Annual Damage', 'A.4_Expected Annual Damage',
       'A.5_Expected Annual Damage', 'A.1_Dike Investment Costs',
       'A.2_Dike Investment Costs', 'A.3_Dike Investment Costs',
       'A.4_Dike Investment Costs', 'A.5_Dike Investment Costs', 'Variance']]

df_ooi["Total Policy Costs"] = df_ooi["RfR Total Costs"]+df_ooi["Expected Evacuation Costs"]+df_ooi["A.1_Dike Investment Costs"]+df_ooi["A.2_Dike Investment Costs"]+df_ooi["A.3_Dike Investment Costs"]+df_ooi["A.4_Dike Investment Costs"]+df_ooi["A.5_Dike Investment Costs"]

df_ooi.head()
```

## Visualization ##

With the retrieved dataframe df_ooi, we can visualize the following KPIs. This now includes total policy costs:

- Expected number of deaths
- Expected annual damage
- Total policy costs
- Variance of costs between locations

The plot below visualizes the correlation between the KPIs per base case and random policy. What we immediately see is that all of the policy values (orange dots) are of a lesser value than the base case values (blue dots). This immediately implies that the every random generated policy performs better than the base case. Therefore, we can imply that every policy that will be implementend (in whatever composition), will help in reducing annual damage and deaths. 

It is however noticeable as well that total policy costs can become quite high in the different random policies: almost 3 billion euros for some policies. It is up to us, to find a policy that is favourable for all actors and with policy costs that are acceptable.

```python
#Now that a combined dataframe has been created, plotting the base case against the random policy runs is now easy.
sns.pairplot(df_ooi, hue='policy', hue_order=["base case", "random"], vars=["Expected Number of Deaths", "Expected Annual Damage", "Total Policy Costs", "Variance"], diag_kind="hist")
plt.legend()
plt.show()
```

As we can see below, the policies (orange) perform better than the base case scenario. Sometimes, the variance is not better distributed, but for most policies it is. Therefore, the aforementioned favourable set-up needs to be found. This will be done in the following notebook "3 - MORO".

Because the figure above is quite information overload for the message we want to convey, the histograms below were created for the report. These are essentially the plots on the diagonal of the figure above, next to each other.
